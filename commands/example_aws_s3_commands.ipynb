# AWS bucket downloading and uploading:

Setup aws bucket credentials

```python
text = '''
[default]
aws_access_key_id = ACCESS_KEY_ID
aws_secret_access_key = SECRET_ACCESS_KEY
region = us-east-2
'''
path = "/content/drive/My Drive/Github/config/awscli.ini"
with open(path, 'w') as f:
   f.write(text)
!cat /content/drive/My\ Drive/Github/config/awscli.ini
```

```python
import os
!export AWS_SHARED_CREDENTIALS_FILE=/content/drive/My\ Drive/Github/config/awscli.ini
path = "/content/drive/My Drive/Github/config/awscli.ini"
os.environ['AWS_SHARED_CREDENTIALS_FILE'] = path
print(os.environ['AWS_SHARED_CREDENTIALS_FILE'])
```
